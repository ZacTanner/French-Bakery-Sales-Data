```python
#Import the required EDA packages
import pandas as pd
import numpy as np
from scipy import stats
```

```python
#Import the dataset
df = pd.read_csv('Bakery.csv')
```

```python
df
```

```python
#Initial EDA exploring sale quantities
df['Quantity'].value_counts()
```

```python
#There were negative values in the above value count. I split off a separarte data frame limited to negative quantities to further exmamine.
NegativeQ = df[df['Quantity'] < 0]
NegativeQ
```

```python
#Negative values could be till errors. They represent less than 1% of all bakery transactions and are not hugely influential. I cleaned by converting to their absolute value.
df['Quantity'] = df['Quantity'].abs()
```

```python
df['Quantity'].value_counts()
```

```python
#In terms of quantity, the number of transactions drops off considerably after 20 items per transaction.
BigOrders = df[df['Quantity'] > 20]
```

```python
BigOrders
```

```python
#The 200 quantity is a massive outlier, so it is dropped.
df = df[df['Quantity'] < 100]
```

```python
df['Quantity'].value_counts()
```

```python
#Converting quantity and ticket number fields to integer and float, respectively, to facilitate data analysis.
df = df.astype({
    'Quantity': 'int',   # Convert 'column1' to integer
    'ticket_number': 'int', # Convert 'column2' to float
})
```

```python
#The index is unhelpful, so dropping it.
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.dtypes
```

```python
#Renaming the fields to simplify analysis.
df = df.rename(columns={
    'article': 'Item',
    'ticket_number': 'Ticket Number',
    'unit_price': 'Unit Price',
    'time': 'Time',
    'date': 'Date'
})
```

```python
df
```

```python
#Reformatting unit price to simplify analysis.
df['Unit Price'] = df['Unit Price'].str.replace(r'[\s€]', '', regex=True)
df['Unit Price'] = df['Unit Price'].str.replace(',', '.', regex=False).astype(float)
```

```python
df
```

```python
df['Unit Price'].value_counts()
```

```python
#Splitting off a separate data frame to examine the most expensive bakery items. Nothing seems unusual or outside the cost of a custom-ordered cake, pie, etc.
expensive = df[df['Unit Price'] > 25]
expensive
```

```python
#Creating a new field that lists total sale price per transaction.
df['Sale Price'] = df['Quantity'] * df['Unit Price']
```

```python
#Export for further EDA using Tableau Public
df.to_csv('Bakery Final.csv', index=False)
```
